```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import path
import netCDF4
# CMOCEAN = oceanographic colorplots: https://matplotlib.org/cmocean/
import cmocean
```

## Load netCDF grid

```python
gname = 'Chesapeake_Bay_NCEI_EasternNeck_handpicked_msl_grd.nc'
nc = netCDF4.Dataset(gname)
```

## Load variables for plotting

```python
#list variables
nc.variables.keys()
lon_rho = nc.variables['lon_rho'][:]
lat_rho = nc.variables['lat_rho'][:]
h = nc.variables['h'][:]
mask = nc.variables['mask_rho'][:]
mask_nan = mask
mask_nan[mask_nan==0]=np.nan
ploth = -1*h*mask_nan
```

## Plot the grid 
add contour lines, colorb

```python
#plot grid
fig,ax=plt.subplots(1,1)
new_cm = cmocean.tools.crop(cmocean.cm.delta,-20,10,0)
plt.pcolormesh(lon_rho,lat_rho,ploth,cmap=new_cm,shading='nearest')
ax.contour(lon_rho,lat_rho,ploth,0,colors='black',linewidths=0.4)
plt.colorbar(label='Depth (m)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.show()
```
