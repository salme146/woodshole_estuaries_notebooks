# Load and plot shapefile

Using Pangeo python environment with earthpy installed
https://pypi.org/project/earthpy/


```python
import os
import matplotlib.pyplot as plt
import geopandas as gpd
import earthpy as et
```

```python
# import shapefile using geopandas
marsh = gpd.read_file('/Users/scook/Desktop/research/proj/Ch_Bay/grid/Marsh_Mask/ENI_NWI_coastal_wetlands_WGS84/ENI_NWI_coastal_wetlands_WGS84.shp')
```

```python
# view  the top 6 lines of attribute table of data
marsh.head(6)
```

```python
# plot the data using geopandas .plot() method
fig, ax = plt.subplots(figsize = (10,10))
marsh.plot(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))

# quickly plot the data adding a legend
marsh.plot(column='WETLAND_TY', 
                         categorical=True, 
                         legend=True, 
                         figsize=(10,6),
                         markersize=45, 
                         cmap="Set2", ax=ax);
```
